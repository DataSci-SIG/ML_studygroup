```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def get_all_flips():
    return np.random.choice([0, 1], size=(10, 1000))

def get_coins():
    all_flips = get_all_flips()
    c1 = all_flips[:, 0]
    crand = all_flips[:, np.random.randint(1000)]
    cmin = all_flips[:, np.argmin(all_flips.sum(axis=0))] 
    return (c1, crand, cmin)

def calc_freq():
    coins = get_coins()
    return [x.sum()/10. for x in coins]
```

```python
experiment = [calc_freq() for _ in range(100000)]
```

```python
c1_means, crand_means, cmin_means = zip(*experiment)
```

```python
plt.hist(c1_means, bins=11)
plt.show()
```

```python
plt.hist(crand_means, bins=11)
plt.show()
```

```python
plt.hist(cmin_means, bins=11)
plt.xlim(0, 1.0)
plt.show()
```
